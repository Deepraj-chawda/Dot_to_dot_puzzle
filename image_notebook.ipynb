```python
import cv2

img = cv2.imread("1.jpg")

edges = cv2.Canny(img,200,400)

cv2.imshow("Detected Edges", edges)

cv2.waitKey(0)
```

```python
my_img = np.full((854, 520, 3),255, dtype = "uint8")
for x,y in c:
    cv2.circle(my_img, (y,x), 1, (0, 0, 0), 2)
    
```

```python
indices = np.where(edges != [0])
indices
```

```python
from PIL import Image
size = max(indices[1])+1,max(indices[0])+1

img = Image.new('RGB',size,color= (255,255,255))
data = img.load()
l1=indices[1].tolist()
l2=indices[0].tolist()
for x,y in zip(l1,l2):
    
    data[x,y] = (0,0,0)
```
